```julia
f(x)  = x^2 # Acá estamos definiendo una funcion!
x = range(-1,2,length=50) # 50 puntos que van desde -1 a 2
y = f.(x) # le aplico la funcion a cada elemento de x
plot(x,y)
scatter!(x,y) # el ! permite graficar sobre el primer grafico
```

```julia
f(x) = x^2 
plot(f,-1,2) # le doy como entrada la funcion el intervalo
```

Opciones para configurar el grafico:
- `color` o `c` : nombre del color en ingles, o #RRGGBB, o RGB(r,g,b)
- `linewidth` o `lw` : espesor de la linea
- `label` o `lab`: texto de la leyenda 
- `marker`o `m` : marcador

```julia
f(x) = exp(-x) * sin(x) + cos(x) * exp(x)
x = range(-2,2,length=100)
y = f.(x)
plot(x,y, color=:red, lw=2, lab= "f", m=:circle)
```
